                               Distribution normale
La distribution normale est l'une des distributions les plus importantes.

Elle est également appelée distribution gaussienne en hommage au mathématicien allemand Carl Friedrich Gauss.

Il correspond à la distribution de probabilité de nombreux événements, par exemple. Scores de QI, rythme cardiaque, etc.

Utilisez la random.normal()méthode pour obtenir une distribution de données normale.

Il comporte trois paramètres :

loc- (Moyen) où se trouve le sommet de la cloche.

scale- (Écart type) à quel point la distribution graphique doit être plate.

size- La forme du tableau renvoyé.

```python
#Générez une distribution normale aléatoire de taille 2x3 :
x = random.normal(size=(2, 3))

print(x)
```

```python
#Générez une distribution normale aléatoire de taille 2x3 avec une moyenne à 1 et un écart type de 2 :
x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
#Visualisation de la distribution normale

sns.distplot(random.normal(size=1000), hist=False)

plt.show()
```

                           Distribution binomiale
La distribution binomiale est une distribution discrète .

Il décrit le résultat de scénarios binaires, par exemple, un tirage au sort, ce sera face ou face.

Il comporte trois paramètres :

n- nombre d'essais.

p- probabilité d'occurrence de chaque essai (par exemple pour un tirage au sort 0,5 chacun).

size- La forme du tableau renvoyé.

```python
#Étant donné que 10 essais de tirage au sort génèrent 10 points de données :

x = random.binomial(n=10, p=0.5, size=10)

print(x)
```

```python
#Visualisation de la distribution binomiale

sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)

plt.show()
```

```python
#Différence entre la distribution normale et binomiale
sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')

plt.show()
```

                                 Distribution de Poisson
La distribution de Poisson est une distribution discrète .

Il estime combien de fois un événement peut se produire dans un temps spécifié. Par exemple, si quelqu'un mange deux fois par jour, quelle est la probabilité qu'il mange trois fois ?

Il possède deux paramètres :

lam- taux ou nombre connu d'occurrences, par exemple 2 pour le problème ci-dessus.

size- La forme du tableau renvoyé.

```python
from numpy import random

x = random.poisson(lam=2, size=10)

print(x)
```

```python
#Visualisation de la distribution de Poisson
sns.distplot(random.poisson(lam=2, size=1000), kde=False)

plt.show()
```

```python
#Différence entre la distribution normale et la distribution de Poisson

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=7, size=1000), hist=False, label='normal')

sns.distplot(random.poisson(lam=50, size=1000), hist=False, label='poisson')

plt.show()
```

```python
#Différence entre la distribution binomiale et la distribution de Poisson
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n=1000, p=0.01, size=1000), hist=False, label='binomial')

sns.distplot(random.poisson(lam=10, size=1000), hist=False, label='poisson')

plt.show()
```

La distribution multinomiale est une généralisation de la distribution binomiale.

Il décrit les résultats des scénarios multinomiaux, contrairement aux scénarios binomiaux où les scénarios ne doivent être que l'un des deux. Par exemple, le groupe sanguin d'une population, le résultat du lancer de dés.

                                    Distribution Zipf
Les distributions Zipf sont utilisées pour échantillonner des données en fonction de la loi de Zipf.

Loi de Zipf : dans une collection, le nième terme commun est 1/n fois le terme le plus courant. Par exemple, le 5ème mot le plus courant en anglais apparaît près de 1/5 fois plus souvent que le mot le plus courant.

Il possède deux paramètres :

a- paramètre de répartition.

size- La forme du tableau renvoyé.

```python
from numpy import random

import matplotlib.pyplot as plt
import seaborn as sns

#Dessinez un échantillon pour la distribution zipf avec le paramètre de distribution 2 et la taille 2x3 :
x = random.zipf(a=2, size=(2, 3))

print(x)

#Visualisation de la distribution Zipf
#Échantillonnez 1 000 points, mais en ne traçant que ceux dont la valeur est < 10 pour un graphique plus significatif

x = random.zipf(a=2, size=1000)
sns.distplot(x[x<10], kde=False)

plt.show()
```

                                 Que sont les ufuncs ?
ufuncs signifie "Universal Functions" et ce sont des fonctions NumPy qui opèrent sur l' ndarrayobjet.

                              Pourquoi utiliser les ufuncs ?
Les ufuncs sont utilisés pour implémenter la vectorisation dans NumPy, ce qui est bien plus rapide que l'itération sur les éléments.

Ils fournissent également la diffusion et des méthodes supplémentaires telles que la réduction, l'accumulation, etc. qui sont très utiles pour le calcul.

Les ufuncs acceptent également des arguments supplémentaires, comme :

where  tableau booléen ou condition définissant où les opérations doivent avoir lieu.

dtype  définir le type de retour des éléments.

out    tableau de sortie où la valeur de retour doit être copiée.
                                 Qu’est-ce que la vectorisation ?
La conversion d'instructions itératives en une opération vectorielle est appelée vectorisation.

C'est plus rapide car les processeurs modernes sont optimisés pour de telles opérations.

                                 Ajouter les éléments de deux listes
liste 1 : [1, 2, 3, 4]

liste 2 : [4, 5, 6, 7]

Une façon de procéder consiste à parcourir les deux listes, puis à additionner chaque élément.

```python
#Sans ufunc, nous pouvons utiliser la méthode intégrée de Pythonzip() :

x = [1, 2, 3, 4]
y = [4, 5, 6, 7]
z = []

for i, j in zip(x, y):
  z.append(i + j)
print(z)
```

```python
#NumPy a une ufunc pour cela, appelée add(x, y) qui produira le même résultat.
import numpy as np

x = [1, 2, 3, 4,3]
y = [4, 5, 6, 7,20]
z = np.add(x, y)

print(z)
```

                         Créez votre propre fonction
                 Comment créer votre propre ufunc
Pour créer votre propre ufunc, vous devez définir une fonction, comme vous le faites avec les fonctions normales en Python, puis vous l'ajoutez à votre bibliothèque ufunc NumPy avec la frompyfunc()méthode.

La frompyfunc()méthode prend les arguments suivants :

function- le nom de la fonction.

inputs- le nombre d'arguments d'entrée (tableaux).

outputs- le nombre de tableaux de sortie.

```python
import numpy as np

#Créez votre propre ufunc pour l'ajouter :
def myadd(x, y):
  return x+y

myadd = np.frompyfunc(myadd, 2, 1)

print(myadd([1, 2, 3, 4], [5, 6, 7, 8]))

print(type(np.add))#verifier le type de ufunc
```

```python
#Utilisez une instruction if pour vérifier si la fonction est un ufunc ou non :
if type(np.add) == np.ufunc:
  print('il est une fonction')
else:
  print("il n'est pas une fonction")
```
